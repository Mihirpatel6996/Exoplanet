```python
df_merged = pd.merge(df_cumulative, df_stellar, on='kepid', how='inner')
```

```python
# Identify columns with missing values
missing_values_count = df_merged.isnull().sum()
missing_values_percent = (missing_values_count / len(df_merged)) * 100
columns_with_missing_values = missing_values_percent[missing_values_percent > 0]

# Print the columns with missing values and their percentages
print("Columns with Missing Values:")
print(columns_with_missing_values)

# Handle missing values in numerical columns using median imputation
numerical_cols_with_missing = df_merged.select_dtypes(include=['number']).columns[df_merged.select_dtypes(include=['number']).isnull().any()]
for col in numerical_cols_with_missing:
  df_merged[col] = df_merged[col].fillna(df_merged[col].median())

# Handle missing values in categorical columns by filling with 'unknown'
categorical_cols_with_missing = df_merged.select_dtypes(include=['object']).columns[df_merged.select_dtypes(include=['object']).isnull().any()]
for col in categorical_cols_with_missing:
  df_merged[col] = df_merged[col].fillna('unknown')
```

```python
# Select the relevant columns
selected_columns = ["kepid", "koi_disposition", "koi_fpflag_nt", "koi_fpflag_ss", "koi_fpflag_co", "koi_fpflag_ec", "koi_period", "koi_prad", "koi_sma", "koi_teq", "koi_insol", "koi_steff", "koi_srad", "koi_smass", "teff", "logg", "feh", "mass", "radius", "dens", "kepmag"]
df_selected = df_merged[selected_columns].copy()

# Handle missing values
for col in df_selected.select_dtypes(include=['number']).columns:
    df_selected[col] = df_selected[col].fillna(df_selected[col].median())

for col in df_selected.select_dtypes(include=['object']).columns:
    df_selected[col] = df_selected[col].fillna('unknown')

# Encode categorical features
df_selected = pd.get_dummies(df_selected, columns=['koi_disposition'], prefix=['koi_disposition'])
```

```python
# Calculate the ratio of koi_prad to radius
df_selected['koi_prad_radius_ratio'] = df_selected['koi_prad'] / df_selected['radius']

# Calculate the ratio of koi_sma to radius
df_selected['koi_sma_radius_ratio'] = df_selected['koi_sma'] / df_selected['radius']

# Calculate the product of koi_insol and koi_teq
df_selected['koi_insol_teq_product'] = df_selected['koi_insol'] * df_selected['koi_teq']
```

```python
from sklearn.model_selection import train_test_split

# Separate features (X) and target (y)
X = df_selected.drop(['koi_disposition_CANDIDATE', 'koi_disposition_CONFIRMED', 'koi_disposition_FALSE POSITIVE'], axis=1)
y = df_selected[['koi_disposition_CANDIDATE', 'koi_disposition_CONFIRMED', 'koi_disposition_FALSE POSITIVE']]

# Split data into training and testing/validation sets
X_train, X_test_val, y_train, y_test_val = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Further split testing/validation sets into testing and validation sets
X_test, X_val, y_test, y_val = train_test_split(
    X_test_val, y_test_val, test_size=0.50, random_state=42, stratify=y_test_val
)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing and validation data using the fitted scaler
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Instantiate a Random Forest classifier
rf_model = RandomForestClassifier()

# Instantiate an XGBoost classifier
xgb_model = XGBClassifier()

# Train the Random Forest model
rf_model.fit(X_train_scaled, y_train)

# Train the XGBoost model
xgb_model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

# Make predictions on the validation set
rf_predictions = rf_model.predict(X_val_scaled)
xgb_predictions = xgb_model.predict(X_val_scaled)

# Calculate evaluation metrics
rf_precision = precision_score(y_val, rf_predictions, average='weighted')
rf_recall = recall_score(y_val, rf_predictions, average='weighted')
rf_f1 = f1_score(y_val, rf_predictions, average='weighted')


rf_accuracy = accuracy_score(y_val, rf_predictions)

xgb_precision = precision_score(y_val, xgb_predictions, average='weighted')
xgb_recall = recall_score(y_val, xgb_predictions, average='weighted')
xgb_f1 = f1_score(y_val, xgb_predictions, average='weighted')
xgb_accuracy = accuracy_score(y_val, xgb_predictions)

# Print the metrics
print("Random Forest Metrics:")
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_f1)
print("Accuracy:", rf_accuracy)

print("\nXGBoost Metrics:")
print("Precision:", xgb_precision)
print("Recall:", xgb_recall)
print("F1 Score:", xgb_f1)
print("Accuracy:", xgb_accuracy)


print(type(X_val_scaled), X_val_scaled.shape)

# Calculate ROC AUC scores
# rf_probs = rf_model.predict_proba(X_val_scaled)
# rf_roc_auc = roc_auc_score(y_val.values.ravel(), rf_probs[:, 1])

# xgb_probs = xgb_model.predict_proba(X_val_scaled)
# xgb_roc_auc = roc_auc_score(y_val.values.ravel(), xgb_probs[:, 1])


# # Print the ROC AUC scores
# print("\nRandom Forest ROC AUC:", rf_roc_auc)
# print("XGBoost ROC AUC:", xgb_roc_auc)


```

```python
from sklearn.model_selection import RandomizedSearchCV

# Define parameter grids for Random Forest and XGBoost models
param_grid_rf = {
    'n_estimators': [50,100,200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}


# Perform Grid Search with Cross-Validation for Random Forest
random_search_rf = RandomizedSearchCV(
    estimator=RandomForestClassifier(),
    param_distributions=param_grid_rf,
    scoring='f1_weighted',
    cv=3,
    n_iter=50,
    n_jobs=-1,
    random_state=42
)
random_search_rf.fit(X_train_scaled, y_train)


```

```python
# Extract Best Model and Parameters for Random Forest
best_rf_model = random_search_rf.best_estimator_
best_rf_params = random_search_rf.best_params_
```

```python
print("Best Random Forest Parameters:", best_rf_params)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Make predictions on the validation set
y_pred = best_rf_model.predict(X_val_scaled)

# Compute evaluation metrics
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred, average='weighted')  # Use 'weighted' for multiclass
recall = recall_score(y_val, y_pred, average='weighted')
f1 = f1_score(y_val, y_pred, average='weighted')

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Print detailed classification report
print("\nClassification Report:\n", classification_report(y_val, y_pred))
```

```python
# Define parameter grid with optimizations
param_grid_xgb = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],  # Reduced range
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2]
}

# Perform RandomizedSearchCV for XGBoost
random_search_xgb = RandomizedSearchCV(
    estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    param_distributions=param_grid_xgb,
    scoring='f1_weighted',
    cv=3,  # Reduced cross-validation folds
    n_iter=50,  # Instead of exhaustive grid search
    n_jobs=-1,
    random_state=42
)


# Fit the model
random_search_xgb.fit(X_train_scaled, y_train)

```

```python
# Extract Best Model and Parameters
best_xgb_model = random_search_xgb.best_estimator_
best_xgb_params = random_search_xgb.best_params_

print("Best XGBoost Parameters:", best_xgb_params)
```

```python
# Make predictions on the validation set
y_pred_xgb = best_xgb_model.predict(X_val_scaled)

# Compute evaluation metrics
accuracy_xgb = accuracy_score(y_val, y_pred_xgb)
precision_xgb = precision_score(y_val, y_pred_xgb, average='weighted')
recall_xgb = recall_score(y_val, y_pred_xgb, average='weighted')
f1_xgb = f1_score(y_val, y_pred_xgb, average='weighted')

# Print the results
print(f"\nXGBoost Model Performance:")
print(f"Accuracy: {accuracy_xgb:.4f}")
print(f"Precision: {precision_xgb:.4f}")
print(f"Recall: {recall_xgb:.4f}")
print(f"F1 Score: {f1_xgb:.4f}")

# Print detailed classification report
print("\nClassification Report:\n", classification_report(y_val, y_pred_xgb))
```
